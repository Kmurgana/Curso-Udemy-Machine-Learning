# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
import random as rm
```

```python
vocals = ["a", "e", "i", "o", "u","ü", ""]
consonants= ["b", "c", "d", "f", "g", "h", "j", "l", "l", "m", "n", "ñ", "p", "q", "r", "s", "t", "v", "w", "x", "y", "z", ""]
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
def random_names(lang, qty):
    lang=lang/2
    names = []
    name = ""
    i = 0
    while (i<qty):      
        for j in range (rm.randrange(lang)):    
                name = name + np.random.choice(vocals)+np.random.choice(consonants)
        if (len(name) != 0):
            name = name.capitalize()
            names.append(name)
            name=""
        else:
            i = i-1   
        i+=1
    return names
```

```python
n = 5000
gender_data = []
income_data = []
for i in range(0,n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
#z -> N(0,1)
#N(m, s) -> m + s * Z
height = 160 + 30 * np.random.rand(n)
weight = 65 + 25 * np.random.rand(n)
age = 30 + 12 * np.random.rand(n)
income = 18000 + 3500 * np.random.rand(n)
```

```python
data = pd.DataFrame(
    {
        "Name": random_names(10, n),
        "Gender": gender_data,
        "Economic Status": income_data,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Income": income       
    }
)
```
