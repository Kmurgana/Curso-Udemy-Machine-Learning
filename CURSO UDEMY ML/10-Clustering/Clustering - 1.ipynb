```python
#Funcin de automatizacin de la matriz de distancias.
def Distance_matriz(df1):
    z=[]
    long = len(df1)
    n=long+1
    d_min_flag=999999
    while(len(df1)!=1):
        for j in (df1.columns.values.tolist()[0:]):
            for k in (df1.index[0:]):
                if (k != j):
                    print("J= ", j, " - K=", k)
                    d_min = round(df1[j][k],1)
                    if (d_min_flag >= d_min):                
                        if (d_min_flag == d_min):
                            print("IF iGUAL--------------")
                            d_min_flag = d_min
                            zz = int(k)
                        else:
                            print("IF3 DIFERENTE------------")
                            d_min_flag = d_min
                            x = int(j)
                            y = int(k)
                            zz = x
        df1[n] = df1[x]+df1[y] 
        df1.loc[n] = df1.loc[x]+df1.loc[y]  
        if(x != zz):
            print("ELIMINACIN CON Z")
            print("X: ", x, " Y: ", y , " Z: ", zz)
            z.append([x,zz,y,d_min_flag,3])    
            for l in df1.columns.values.tolist():
                df1.loc[n][l] = min(df1.loc[x][l], df1.loc[zz][l], df1.loc[y][l])
                df1.loc[l][n] = min(df1.loc[l][x], df1.loc[l][zz], df1.loc[l][y])
            df2 = df1.drop([x,y,zz])
            df1 = df2
            df2 = df1.drop([x,y,zz], axis = 1)
            df1 = df2
        elif(x == zz):
            print("ELIMINACIN SIN Z")
            print("X: ", x, " Y: ", y , " Z: ", zz)
            z.append([x,y,d_min_flag,2])  
            for l in df1.columns.values.tolist():
                df1.loc[n][l] = min(df1.loc[y][l], df1.loc[x][l])
                df1.loc[l][n] = min(df1.loc[l][y], df1.loc[l][x])
            df2 = df1.drop([x,y])
            df1 = df2
            df2 = df1.drop([x,y], axis = 1) 
            df1 = df2
        print ("Distancia mnima: ", d_min_flag)
        print ("")
        print(df2)
        d_min_flag = 999999
        n +=1
        print("-------------------------------------------------------------")           
    return (df2,z)
```

# Clustering Jerrquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward") #El mtodo ward minimiza los cuadrados de las varianzas, es el mtodo ms raro
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el CLustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90.0, leaf_font_size=15)
plt.show()
```

```python
Z = linkage(data[movies], "average") #El mtodo average es de promedio
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el CLustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90.0, leaf_font_size=15)
plt.show()
```

```python
Z = linkage(data[movies], "complete") # El enlace es muy similar al de ward
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el CLustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90.0, leaf_font_size=15)
plt.show()
```

```python
Z = linkage(data[movies], "single") # El enlace single es el simple
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el CLustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90.0, leaf_font_size=15)
plt.show()
```
