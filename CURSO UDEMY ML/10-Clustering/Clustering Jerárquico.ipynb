# Clustering jerárquico y dendrogramas

Antes de empezar, pongamos un poco de notación para hablar todos en el mismo idioma
* X dataset (array de n x m) de puntos a clusterizar
* n número de datos
* m número de rasgos
* Z array de enlace de cluster con la información de las uniones
* k número de clusters

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]], size = [100,]) #generamos 100 datos diferentes
b = np.random.multivariate_normal([0,20],[[3,1],[1,4]], size = [50,]) #generamos 50 datos diferentes 
#El primer [] significa media en X y en y por eso en a, sería 10 en x y 0 en y - b es 0 en x y 20 en y
#El segundo [] indica la desviación tipica en horizontal (3 puntos para ambos)
#El tercer [] indica la desviación típica en vertical (4 puntos para ambos)
#El cuarto [] indica el tamaño de la muestra
X = np.concatenate((a,b))
print(X.shape)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
Z = linkage(X, "ward")
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
c, coph_dist = cophenet(Z, pdist(X))
c
```

El coeficiente de **cophenet** sirve para saber el valor de presición de las distancias calculadas por el método linkage ward comparando con las distancias en el data set original, o sea en X. En este caso nos arrojó una distancia de 0.98 lo que quiere decir un 98% de presición o de conservación de distancias con respecto a los valores originales con los clústers que se han generado.

### Representación gráfica de un dendrograma

```python
plt.figure(figsize=(20,10))
plt.title("Dendrograma del clustering jerárquico")
plt.xlabel("Índices de la Muestra")
plt.ylabel("Distancias")
dendrogram(Z, leaf_rotation=90., leaf_font_size=8.0 )
plt.show()
```

```python
Z[-4:,] # Vemos los últimos cuatro elementos, el último elemento tiene distancia 180.
```

Se puede ver claramente la segmentación de los dos cluster en **anaranjado** están los 50 elementos del b, en **verde** los otros 100 elementos. Nuestra población está dividida en dos grupos.

Se pueden modificar los coleres del dendrograma, con la función color_threshold = **%** este porcentaje es equivale al último valor de distancia, o sea sería en este caso = 0.7 * 180

```python
plt.figure(figsize=(20,10))
plt.title("Dendrograma del clustering jerárquico")
plt.xlabel("Índices de la Muestra")
plt.ylabel("Distancias")
dendrogram(Z, leaf_rotation=90., leaf_font_size=8.0, color_threshold=0.1*180 )
plt.show()
```

Por lo que vemos, gráficamente con colores es más fácil de ver los clusters principales.

### Truncar el dendrograma

```python
plt.figure(figsize=(10,10))
plt.title("Dendrograma del clustering jerárquico truncado")
plt.xlabel("Índices de la Muestra")
plt.ylabel("Distancias")
dendrogram(Z, leaf_rotation=90., leaf_font_size=12.0, color_threshold=0.7*180,
          truncate_mode="lastp", p=12, show_leaf_counts=True, show_contracted=True)
plt.show()
```

Como muestra las últimas 12 uniones, y sólo el 40 es el único elemento que quedó sólo por unir, entonces es el único que se le puede ver el label que se une al resto de los clústers. Los que tienen la bolita negra más arrriba quiere decir que se han unido más tarde. Si *show_leaf_counts* se pone en true, se muestra la canitdad de elemntos entre () de la cantidad de elementos que pertenecen a cada clúster.

### Dendrograma Tuneado

```python
def dendrogram_tuned(*args, **kwargs):
    max_d=kwargs.pop("max_d", None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)
    
    ddata = dendrogram(*args, **kwargs)
    
    if not kwargs.get('no_plot', False):
        plt.title("Clustering Jerárquico con Dendrograma Truncado")
        plt.xlabel("Índice del Dataset (o tamaño del cluster)")
        plt.ylabel("Distancia")
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y>annotate_above: 
                plt.plot(x,y, 'o', c=c)
                plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),
                            textcoords="offset points", va="top", ha="center")
    if max_d:
        plt.axhline(y=max_d, c='k')
        
    return ddata
```

```python
plt.figure(figsize=(10,10))
dendrogram_tuned(Z, truncate_mode='lastp', p=12, 
                 leaf_rotation=90., leaf_font_size=12., show_contracted=True,
                 annotate_above=10, max_d = 20)
plt.show()
```

### Corte automático del dendrograma

$ Inconsistency_i= \frac{h_i- avg(h_i)}{std(h_j)} $

```python
from scipy.cluster.hierarchy import inconsistent
```

```python
depth=3
incons = inconsistent(Z, depth)
incons[-10:]
```

En particular, el método de la inconsistencia depende muchísimo del grado de profundidad **depth** que nosotros le asignemos, por lo que en el momento que nosotros lo coloquemos un valor muy chico, la incosistencia será muy parecida en todos los casos, por lo que este método no es bueno para determinar el corte que se debe realizar en el cluster, o sea, el nivel.

### Método del codo

```python
last = Z[-10:,2]
last_rev = last[::-1] # Revertimos los valores, el -1 los dá vuelta
print(last_rev)
idx = np.arange(1, len(last)+1)
plt.plot(idx, last_rev)
aceleracion = np.diff(last,2) #De dos en dos va a ir restando o calculando la aceleración
aceleracion_rev = aceleracion[::-1]
plt.plot(idx[:-2]+1, aceleracion_rev)
plt.show()
k = aceleracion_rev.argmax() +2
print("El número óptimo de cluster es %s" %str(k))
```

Haciendo ésto, podemos observar cómo el último elemento genera un salto, una aceleración bastante elebada con respecto al resto de los saltos, o de uniones de clústerings. En **naranja** vemos la derivada de los saltos, vemos que entre el primero y el segundo hay un salto de 150. En **azul** vemos lo que son los saltos netos entre cluster y clúster. El problema de éste método es que se necesita al menos dos cluster, desde k=2 hasta k=9, la técnica del codo no sirve si queremos averiguar si todos pueden vivir en el mismo clúster.

## Recuperar los clusters y sus elementos

```python
from scipy.cluster.hierarchy import fcluster
```

```python
max_d = 20
clusters = fcluster(Z, max_d, criterion="distance")
clusters
```

Con este método, vemos que tenemos 150 elementos en el array, que dicen 1,2 y 3 respectivamente que significa a qué cluster pertenece cada elemento. Como vemos, **max_d** es el valor de umbral que le colocamos para que realice el corte y separe los clústers, hacia abajo, entonces viendo el gráfico de Z, vemos que abajo del nivel 20 tenemos 3 clústers, por eso lo separa en tres.

```python
plt.figure(figsize=(10,10))
dendrogram_tuned(Z, truncate_mode='lastp', p=12, 
                 leaf_rotation=90., leaf_font_size=12., show_contracted=True,
                 annotate_above=10, max_d = 20)
plt.show()
```

```python
k=3
clusters = fcluster(Z, k, criterion="maxclust")
clusters
```

En este método, el resultado es exactamente el mismo que con máx_d, pues especificamos la cantidad de clústers con K=3. Son dos formas, conociendo la altura a la que quiero cortar el dendrograma y/o conociendo la cantidad de clústers que quiero contar.

```python
fcluster(Z, 8, depth=10)
```

Como vemos de esta manera es como el método de la incositencia, que lee los últimos diez clústers generados para poder ver claramente cuáles son los clusters correspondientes.
