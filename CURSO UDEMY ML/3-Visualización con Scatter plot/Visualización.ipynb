# Plots y visualizaci√≥n de los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mainpath = "C:/Users/kmurg/OneDrive/Escritorio/Kev/1-Cursos/Machine Learning/python-ml-course-master/datasets"
data = pd.read_csv(mainpath+"/"+"customer-churn-model/Customer Churn Model.txt")
```

```python
%matplotlib inline
```

## Scatter Plot

```python
data.plot(kind="scatter", x="Day Mins", y ="Day Charge")
```

```python
data.plot(kind="scatter", x="Night Mins", y ="Night Charge")
```

```python
figure, axs= plt.subplots(2,2, sharey=True, sharex=True)
data.plot(kind="scatter", x="Day Mins", y="Day Charge", ax=axs[0][0])
data.plot(kind="scatter", x="Night Mins", y ="Night Charge", ax=axs[0][1])
data.plot(kind="scatter", x="Day Calls", y="Day Charge", ax=axs[1][0])
data.plot(kind="scatter", x="Night Calls", y ="Night Charge", ax=axs[1][1])
```

## Histogramas de frecuencias

```python
k = int(np.ceil(1+np.log2(3333)))    #Redondeamos hacia arriba el conchasumadre dato floor es hacia abajo
plt.hist(data["Day Calls"], bins = k )
plt.xlabel("N√∫mero de llamadas al d√≠a")
plt.ylabel("Frecuencia")
plt.title("Histograma de n√∫mero de llamadas al d√≠a")
```

```python
plt.boxplot(data["Day Calls"])
plt.ylabel("N√∫mero de llamadas diarias")
plt.title("Llamadas diarias")
```
