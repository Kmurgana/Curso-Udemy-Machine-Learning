# Carga de datos a través de la función open

```python
data3= open (mainpath+"/"+"customer-churn-model/Customer Churn Model.txt", 'r') #Abrimos el archivo en modo escritura
```

```python
data3.seek(0)
cols = data3.readline().strip().split(",") #Strip elimina los espacios en blanco adelante y al final de la línea, Split divide el framgento por un delimitador
n_cols = len(cols) #Obtenemos el número de columnas
```

```python
counter = 0
main_dict = {}   #Definimos un diccionario vacío
for col in cols:
    main_dict[col] = []  #Los nombres de la columna son las claves de los diccionarios
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter +=1
print ("El data set tiene %d filas y %d columnas" % (counter, n_cols))
#Este método es un poco menos ortodoxo ya que tenemos que crear dos ciclos for y más varables y aparte no cerramos el archivo
```

```python
df3 = pd.DataFrame(main_dict) #Creamos el data frame
df3.head()
```

```python
import urllib3
import csv
import os
```

```python
#Creamos una función que reciba de parámetro la página web, el archivo, separador, delimitador y encoding.
def downloadfromURL(url, filename, sep = ",", delim = "\n", encodig='utf-8'):  

    #Primero importamos la librería y hacemos la conexión con la web de los datos
    http = urllib3.PoolManager()
    r = http.request('GET', url)

    #Un estado de 200 indica que el servidor está respondiendo adecuadamente
    print ("El estado de la respuesta es %d " %(r.status))
    response = r.data

    #El objeto response contiene un string binario, así que lo convertimos a un string decodificándolo en UTF-8
    str_data = response.decode(encodig)

    #Dividimos el string en un array de filas, separándolo por intros
    lines = str_data.split(delim)

    #Laprimera línea contiene la cabecera, así que la extraemos
    col_names = lines[0].split(sep)
    n_cols  = len (col_names)

    counter_medals = 0
    medals_dict = {}

    #Creamos diccionario vacío con el mismo número y nombre de columnas que col_names
    for col in col_names:
        medals_dict[col] = []

    for line in lines:
        if (counter_medals>0 and len(line)>0): #Se le coloca la condición len(line)>0 para evitar romper el dataframe
            values = line.strip().split(sep)
            for i in range(n_cols):
                medals_dict[col_names[i]].append(values[i])
        counter_medals += 1

    print ("El data set tiene %d filas y %d columnas"%(counter_medals, n_cols))

    df5 = pd.DataFrame(medals_dict)
    
    mainpath = "C:/Users/kmurg/OneDrive/Escritorio/Kev/1-Cursos/Machine Learning/python-ml-course-master/datasets"
    #filename = "titanic/medals."
    fullpath= os.path.join(mainpath, filename)

    df5.to_csv(fullpath+"csv")
    df5.to_excel(fullpath+"xlsx")
    print("Los ficheros se han guardado correctamente en " +fullpath )
    
    return df5
```

```python
medals_df = downloadfromURL(medals_url, "titanic/medals.", sep=",", delim="\n", encodig='utf-8')
medals_df.head()
```
